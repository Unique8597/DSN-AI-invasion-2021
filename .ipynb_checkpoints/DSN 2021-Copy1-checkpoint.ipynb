```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, StandardScaler, Normalizer
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import f1_score,classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Importing my dataset

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
sample_submission = pd.read_csv('sample_submission.csv')
```

```python
train.head()
```

#### Counting values of train and test

```python
train.subscribed.value_counts()
```

```python
train.shape
```

```python
test.shape
```

#### getting number columns and categorical columns

```python
no_cols = train.describe().columns
```

```python
len(no_cols)
```

```python
all_columns  =  test.columns
cat_col = [col for col in all_columns if col not in no_cols]
```

```python
del cat_col[0]
```

#### Converting categorical columns

```python
for cat in cat_col:
    lb = LabelEncoder().fit(train[cat])
    train[cat] = lb.transform(train[cat])
    test[cat] = lb.transform(test[cat])
```

#### Dropping unuseful data columns

```python
X = train.drop('customer_id', axis = 1)
X = X.drop('contact', axis = 1)
test = test.drop('customer_id', axis = 1)
test = test.drop('contact', axis = 1)
```

```python
X.head(10)
```

#### Scaling train dataset

```python
stad = MinMaxScaler()
stad.fit(X)
new_train = stad.transform(X)
```

#### Dropping target column

```python
y_target = X['subscribed']
```

#### Train test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y_target, test_size = 0.3, random_state = 0)
```

```python
X_train.head(30)
```

```python
shuffled_df = X_train.sample(frac = 1, random_state = 0)
```

```python
subscribed_df = shuffled_df.loc[shuffled_df['subscribed']==1]
```

```python
n_subscribed_df = shuffled_df.loc[shuffled_df['subscribed']==0].sample(n= 12000, random_state = 42)
```

```python
normalized_df = pd.concat([subscribed_df, n_subscribed_df])
```

```python
new_X = normalized_df.drop('subscribed', axis = 1)
```

```python
new_X.head()
```

```python
plt.figure(figsize = (6,6))
sns.countplot('subscribed', data = normalized_df)
plt.show()
```

#### Model and fitting

```python
reg = GradientBoostingClassifier(random_state = 42)
reg.fit(new_X, Y_train)
```

#### Prediction

```python
new_reg = reg.predict(X_test)
```

```python
pd.Series(new_reg).value_counts()
```

#### f1 score

```python
f1_score(Y_test, new_reg)
```

```python
print(classification_report(Y_test, new_reg))
```

#### main test

```python
new_reg_2 = reg.predict(test)
```

#### Submission

```python
sample_submission.subscribed = new_reg_2
```
